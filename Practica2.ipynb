Plottin Sin Linea.
Para trazar solko marcadores, puedes utilizar notacion de la cadena de acceso directo 'o', que significa "anillos".
------
Ejemplo: Dibuja dos puntos en el diagrama, uno de la posicion (1,3) y otro en la posicion de (8,10)
-----------

```python
import matplotlib.pyplot as plt
import numpy as np

puntosX=np.array([1,8])
puntosY=np.array([3,10])

plt.plot(puntosX, puntosY, 'o')
plt.show()
```

Diagramas de dispersion (scatter)
Se emplea para mostar la relacion entre dos variables numericas, detectando correlaciones o agrupamientos.
-------
Ejemplo
-----

```python
import matplotlib.pyplot as plt

altura=[150, 160, 170, 180, 190, 200]
peso=[50, 60, 70, 80, 85, 95]

plt.scatter(altura, peso, color="red", marker= "X")
plt.title("relacion entre la altura y el peso")
plt.xlabel("altura (cms)")
plt.ylabel("peso (kg)")
plt.show()
```

Graficos de barras(bar y barh)
Sirven para comparar categorias o variables tanto en orientacion vertical como horizontal.
-----
Ejemplo de barras verticales:
-------

```python
import matplotlib.pyplot as plt

categorias=["A", "B", "C", "D",]
valores=[4, 7, 1, 8]

plt.bar(categorias, valores, color="orange")
plt.title("Graficas de barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
```
