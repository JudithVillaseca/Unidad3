Conceptos Basicos de la Libreria Matplotlib
Matplotlib es una libreria de trazados de graficas de bajo nivel en Python que sirve como una herramienta de visualizacion. Matplotlib es de codigo abierto y podemos utilizarlo abiertamente. Matplotlib esta escrito principalmente en Python con algunos segmentos C.
-------------------------
Instalacion de Matplotlib
Si tienes Python y Pip ya instalado en tu sistema instala matplotlib utilizando este comando:
pip install matplotlib
Desde la linea de comandos.
-----------------------
Importar Matplotlib
Una ves instalado matplotlib importalo en tus aplicaciones agregando el modulo import ejemplo:
import matplotlib
------------------------
Matplotlib Pyplot
La mayoria de las herramientas de matplotlib se encuentran dentro del submodulo pyplot y normalmente se importan bajo el alias plt:
import matplotlib.pyplot as plt
Ahora el paquete pyplot puede denominarse PLT.
